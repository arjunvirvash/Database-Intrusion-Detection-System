```python
import pandas as pd
import random

# Read your CSV file
df = pd.read_csv('/kaggle/input/rnp-miner-dataset/normal.csv')

# Set seed for reproducibility
random.seed(42)

# Randomly select 5000 row indices
total_rows = len(df)
selected_indices = random.sample(range(total_rows), 5000)

# Create the two datasets
selected_df = df.iloc[selected_indices].copy()
remaining_df = df.drop(index=selected_indices).copy()

# Save to CSV
selected_df.to_csv('testing_queries.csv', index=False)
remaining_df.to_csv('training_queries.csv', index=False)
```

```python
import pandas as pd

# Load existing selected queries and new malicious dataset
existing_selected = pd.read_csv('/kaggle/working/testing_queries.csv')
new_malicious = pd.read_csv('/kaggle/input/rnp-miner-dataset/malicious.csv')  # Your 1255-query dataset

# Randomly select 500 samples from new dataset (without replacement)
selected_new = new_malicious.sample(n=500, random_state=42)

# Combine datasets
combined_selected = pd.concat([existing_selected, selected_new], ignore_index=True)

# Save updated datasets
combined_selected.to_csv('/kaggle/working/testing_queries.csv', index=False)
selected_new.to_csv('newly_added_malicious_queries.csv', index=False)

# Optional: Save remaining 755 queries from new dataset
remaining_new = new_malicious.drop(selected_new.index)
remaining_new.to_csv('remaining_malicious_queries.csv', index=False)
```
